##  1.  Introducation 



The Internet’s history goes back some decades by now – email has been around since the 1960s, file sharing since at least the 1970s, and TCP/IP was standardized in 1982.

But it was the creation of the World Wide Web in 1989 that revolutionized our history of communication. But across most of the world, the internet had not yet had much influence – 93% in the East Asia and Pacific region and 99% in South Asia and in Sub-Saharan Africa were still offline in 2000. At the time of the Dot-com-crash less than 7% of the world was online.

Fifteen years later, in 2016, three-quarters (76%) of people in the US were online and during these years countries from many parts of the world caught up: in Malaysia 79% used the internet; in Spain and Singapore 81%; in France 86%; in South Korea and Japan 93%; in Denmark and Norway 97%; and Iceland tops the ranking with 98% of the population online.2

At the other end of the spectrum, there are still countries where almost nothing has changed since 1990. In the very poorest countries – including Eritrea, Somalia, Guinea-Bissau, the Central African Republic, Niger, and Madagascar – fewer than 5% are online. And at the very bottom is North Korea, where the country’s oppressive regime restricts the access to the walled-off North Korean intranet Kwangmyong and access to the global internet is only granted to a very small elite.

But the overarching trend globally – and, as the chart shows, in all world regions – is clear: more and more people are online every year. The speed with which the world is changing is incredibly fast. On any day in the last 5 years there were on average 640,000 people online for the first time.3

This was 27,000 every hour.

For those who are online most days it is easy to forget how young the internet still is. The timeline below the chart reminds you how recent websites and technologies became available that are integrated to the everyday lives of millions: In the 1990s there was no Wikipedia, Twitter launched in 2006, and Our World in Data is only 4 years old (and look how many people have joined since then4).

And while many of us cannot imagine their lives without the services that the internet provides, the key message for me from this overview of the global history of the internet is that we are still in the very early stages of the internet. It was only in 2017 that half of the world population was online; and in 2018 it is therefore still the case that close to half of the world population is not using the internet.5

### 3.1 World internet usage 1990-2017 and Subscription 2004-2020

```python
# figure size
plt.subplots(figsize=(20, 5))
# world inetrnet usage 1900-2017
df_int=internet[internet['Entity']=='World']
# world subscription 2001-2020
df_sub=broadband[broadband['Entity']=='World']
# plot lineplot x axis is year and y axis is internet usage, color is green
plt.plot(df_int['Year'], df_int['Internet_Usage'], color='g', label='usage')
# plot lineplot x axis is year and y axis is subscription, color is red
plt.plot(df_sub['Year'], df_sub['Broadband_Subscriptions'], color='r', label='subscription ')
# y axis label
# title of plot
plt.title('world Internet_Usage, subscription  vs Year')
plt.legend()
# to see grph
plt.show()
```

##### 3.2 What are the top 5 countries with the highest internet use (by population share)?

```python
# group data by Entity and sum up then sort value by descending order
highest_internet_use_country=internet.groupby('Entity')['Internet_Usage'].sum().sort_values(ascending=False)
highest_internet_use_country.head()
```

The top five internet uses country are European,  because the data dived for total population so those country have small population size but almost 100% uses internet

##### 3.3. How many people had internet access in those countries in 2019?

```python
# create list of five country
name_country = ['Norway', 'Iceland', 'Denmark', 'Netherlands', 'Sweden']
# slice five country from people table
country_rows=people['Entity'].isin(name_country)
# filter only 2019 year data
year_2019=people['Year']==2019
# get data
five_country_2019=people[country_rows &year_2019][['Entity', 'Users']]
# sort data by users in descending order
five_country_2019=five_country_2019.sort_values(by='Users', ascending=False)
five_country_2019['per million'] = five_country_2019['Users']/1000000
# take a look population 
five_country_2019.head()

```

##### total population of top five internet uses country

```python
# total population that used internet in those country by million
five_country_2019['Users'].sum()/ 1000000
```

##### 3.4 What are the top 5 countries with the highest internet use for each of the following regions: 'Africa Eastern and Southern', 'Africa Western and Central', 'Latin America & Caribbean', 'East Asia & Pacific', 'South Asia', 'North America', 'European Union'?

```python
# list of region
list_region=['Africa Eastern and Southern', 'Africa Western and Central', 'Latin America & Caribbean', 'East Asia & Pacific', 'South Asia', 'North America', 'European Union']
# slice those region from data table
df_region=internet[internet['Entity'].isin(list_region)]
# group by entity and sum internet usage then sort values in descending order
five_region=df_region.groupby('Entity')['Internet_Usage'].sum().sort_values(ascending=False)[:5]
# take a look top five region
five_region.head()
```

##### 3.5 Create a visualization for those five regions' internet usage over time.

```python
# index of data on list
list_five_region=five_region.index.tolist()
# value of data
value_five_region=five_region.values
# set size of figure width 20 and height 7
plt.subplots(figsize=(20, 7))
# plot bar plot x axis is name of region and y axis is internet usage
sns.barplot(x=list_five_region, y=value_five_region)
# set y label 🏷 
plt.ylabel('internet usage')
# set title of figure
plt.title('internet usage of top five region')
# see figure
plt.show()
```

##### 3.6 What are the 5 countries with the most internet users?

```python
# drop region because region code column is null so we can remove by this way 
df_country=people.dropna()
# filter only 2020 users and remove world rows from entity column
df_users=df_country[(df_country['Year']==2020) & (df_country['Entity']!='World')]
# sort values by descending order
df_users=df_users.sort_values(by='Users', ascending =False)[:5]
df_users['users per million']=df_users['Users']/1000000
# take a look top five internet users country
df_users.head()
```

##### how many billion internet users is there in those top five internet users country ?

```python
total_users=df_users['Users'].sum()
print('there is',total_users/ 1000000000 , 'billion')
```

##### how many percent cover top five country from world internet users ?

```python
df_world=people[(people['Entity']=="World") & (people['Year']==2020)]
coverage_five=float(total_users/df_world['Users'])
print(coverage_five*100, 'percent')
```

###### 3.7. What is the correlation between internet usage (population share) and broadband subscriptions for 2019?

```python
# filter only 2019 year  internet data
df_internet_usage=internet[internet['Year']==2019]
# filter only 2019 year  broadband data
df_broadband_2019=broadband[broadband['Year']==2019]
# we need to merge because it is not the same rows of both data
df_merge=pd.merge(df_internet_usage, df_broadband_2019, how='left', on='Code')
# remove null value 
df_merge=df_merge.dropna()
# filter columns
df_2019=df_merge[['Entity_x', 'Code', 'Internet_Usage', 'Broadband_Subscriptions']]
# range of x axis
x=range(len(df_2019))
# figure size
plt.subplots(figsize=(20, 7))
# plot internet usage 
plt.plot(x, df_2019['Internet_Usage'])
# plot broadband subscription 
plt.plot(x, df_2019['Broadband_Subscriptions'])
# label
plt.legend(['uses', 'subscrib'])
plt.show()
```

###### there is strong positive relationship between internet usage and broadband subscription

```python
# correlation internet use and broadband subscription 
df_2019.corr()
```

##### this table show us, the internet usage and broadband subscription are 0.55 relationship.
